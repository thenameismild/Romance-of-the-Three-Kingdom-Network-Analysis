```python
import nltk
nltk.download('vader_lexicon')
import pandas as pd
import numpy as np
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
from time import sleep
```

```python
#adding characters to list
characters = []
infile = open('characters.txt','r')


for line in infile:
    characters.append(line.strip().split('\n'))


infile.close()

print(characters)

for i in range(0,len(characters)):
  characters[i] = characters[i][0]
```

```python
#sentences list
import re

with open('fulltext.txt', 'r') as file:
  data = file.read().replace('\n', ' ')
  text = re.sub('[^A-Za-z0-9.]+', ' ', data)

sentences = text.split('.')
sentences.pop(293)
```

```python
sentiment = pd.DataFrame(sentences, columns=['Sentence'])
sentiment['Sentiment'] = range(len(sentences))

i = 0
for sentence in sentences:
    scores = sid.polarity_scores(sentence)
    sentiment.loc[i,'Sentiment'] = float(scores['compound'])
    i += 1

print(sentiment)
```
